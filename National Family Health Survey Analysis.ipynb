# **National Family Health Survey (NFHS): Health & Nutrition Analysis Across Indian Districts**

## **INTRODUCTION**
Healthcare and nutrition are critical indicators of a country's social and economic development. Understanding the factors that influence public health outcomes helps policymakers, healthcare professionals, and researchers design effective interventions to improve the well-being of the population.

The National Family Health Survey (NFHS) is a large-scale, multi-round survey conducted across India that provides comprehensive information on population health, nutrition, maternal and child healthcare, family welfare, and lifestyle indicators. The survey collects district-level and state-level data, enabling detailed analysis of regional variations and health disparities.

In this project, the NFHS dataset is analyzed to explore key health and demographic indicators across different districts and states of India. The analysis focuses on areas such as female literacy, maternal healthcare, child nutrition, vaccination coverage, anemia prevalence, and healthcare accessibility. Through data cleaning, exploratory data analysis (EDA), statistical summaries, and visualizations, the project aims to identify meaningful patterns, trends, and relationships within the data.

The insights derived from this analysis can contribute to a better understanding of public health challenges and support evidence-based decision-making for improving healthcare outcomes across the country.

## **DATASET DESCRIPTION**

The dataset used in this project is derived from the **National Family Health Survey (NFHS)**, one of India's most comprehensive health and demographic surveys. The dataset was downloaded in CSV format from the India Data Portal and contains district-level indicators collected under the National Family Health Survey. The dataset contains **1,267 records** and **110 features**, covering multiple health and social indicators from different regions of the country. The data includes information from NFHS survey rounds conducted during **2014–15** and **2019–20**, enabling the analysis of health conditions and development trends over time.

The dataset consists of both **categorical variables** (such as state names, district names, and survey year) and **numerical variables** representing percentages, rates, and health-related measurements. These variables provide insights into healthcare accessibility, maternal and child health, nutritional status, literacy levels, and lifestyle habits.
### <u>Description of Important Indicators</u>
Some column names in the dataset are abbreviated. The following descriptions help in understanding the key indicators used in the analysis:
* **fem_literacy** – Female literacy rate (%)
* **child_5_stunted** – Percentage of children under 5 years who are stunted (low height for age)
* **child_5_wasted** – Percentage of children under 5 years who are wasted (low weight for height)
* **child_5_underweight** – Percentage of children under 5 years who are underweight
* **cv_12_23_full_vacc** – Percentage of children aged 12–23 months who received full vaccination
* **mc_anc_4** – Percentage of mothers who received at least four antenatal care (ANC) visits during pregnancy
* **dc_insti_births** – Percentage of institutional births (deliveries occurring in healthcare facilities)
* **wom_15_49_anaemic** – Percentage of women aged 15–49 years suffering from anemia
* **men_15_49_anaemic** – Percentage of men aged 15–49 years suffering from anemia
* **child_6_59_anaemic** – Percentage of children aged 6–59 months suffering from anemia
* **hh_improved_sanitation** – Percentage of households with improved sanitation facilities
* **hh_electricity** – Percentage of households having electricity access
* **hh_clean_fuel** – Percentage of households using clean cooking fuel
* **wom_own_mobile** – Percentage of women who own and use a mobile phone
* **wom_bank_account** – Percentage of women having a bank or savings account
* **tobacco_use** – Percentage of individuals using tobacco products
* **alcohol_use** – Percentage of individuals consuming alcohol

## **DATA PREPROCESSING**

### **<u>MISSING VALUES CHECK</u>**

```python
missing_values = df.isnull().sum()
```

### **<u>FILLING MISSING VALUES</u>**

```python
# Fill numerical columns with median
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
```

#### ***<u>Observation</u>***

***Missing numerical values were replaced using the median because the dataset contains skewed distributions and the median is less affected by outliers than the mean.***

```python
# Fill categorical columns with mode
cat_cols = df.select_dtypes(include=['string']).columns
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

### **<u>DERIVED COLUMNS</u>**
#### **1. Extract Survey Year**

```python
df = df.copy()
df['survey_year'] = df['year'].astype(str).str[:4]
df[['year','survey_year']].head()
```

***A copy of the original DataFrame was first created using df.copy() to prevent any modifications from affecting the source dataset during preprocessing. The values in the year column were then converted to string format using astype(str) to enable string operations. The first four characters representing the starting year of each survey period were extracted using .str[:4] and stored in a new column named survey_year. This transformation converted survey periods such as '2015-16' and '2019-20' into a single year representation, simplifying year-wise grouping, sorting, trend analysis, and visualization. Finally, the head() function was used to display the first five rows of the original year and newly created survey_year columns to verify that the extraction had been performed correctly.***

#### **2.Female Literacy Category**

```python
df['literacy_category'] = pd.cut(
    df['fem_literacy'],
    bins=[0,50,75,100],
    labels=['Low','Medium','High']
)
df['literacy_category'].value_counts(dropna=False)
```

***The continuous female literacy rate variable was transformed into a categorical variable using pd.cut(). Literacy rates were grouped into Low (0–50%), Medium (50–75%), and High (75–100%) categories and stored in a new column named literacy_category. This categorization simplified the analysis of literacy patterns across regions and enabled easier comparison of records based on literacy levels. The distribution of records across categories was then examined using value_counts(), with dropna=False ensuring that any missing values were also identified and accounted for during the analysis.***

#### **3.Healthcare Access Category**

```python
df['healthcare_access'] = pd.cut(
    df['dc_insti_births'],
    bins=[0,50,80,100],
    labels=['Poor','Moderate','Good']
)
df['healthcare_access'].value_counts(dropna=False)
```

***The institutional birth rate (dc_insti_births) was transformed into a categorical variable named healthcare_access using pd.cut(). Records were classified into Poor (0–50%), Moderate (50–80%), and Good (80–100%) healthcare access categories based on institutional birth coverage. This categorization simplified the assessment of healthcare accessibility across regions and facilitated comparative analysis. The distribution of records across these categories was then examined using value_counts(), with dropna=False ensuring that any missing values were also identified and included in the summary.***

#### **4.Nutrition Risk Category**

```python
df['nutrition_risk'] = pd.cut(
    df['child_5_stunted'],
    bins=[0,20,35,100],
    labels=['Low','Medium','High']
)
df['nutrition_risk'].value_counts(dropna=False)
```

***The child stunting rate (child_5_stunted) was transformed into a categorical variable named nutrition_risk using pd.cut(). Records were classified into Low (0–20%), Medium (20–35%), and High (35–100%) nutrition risk categories based on the prevalence of stunting among children under five years of age. This categorization simplified the assessment of nutritional conditions across regions and supported comparative analysis of malnutrition levels. The distribution of records across the categories was then examined using value_counts(), with dropna=False ensuring that any missing values were also identified and included in the summary.***
